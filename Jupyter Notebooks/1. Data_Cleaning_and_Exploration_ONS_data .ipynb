# Data Cleaning and Exploration

# Girls names data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create two dataframes, one for Girls and one for Boys by selecting the correct sheet name from the spreadsheet

girls_data = pd.read_excel(r'../data/babynames1996to2020.xls', sheet_name='Girls')

girls_data #displays preview of girls data
```

## Start data clean by looking for unnecessary rows/columns

```python
# take a closer look at the tail of the data to see which rows may need removing
girls_data.tail(10)
```

```python
# Need to remove the bottom 7 rows as they do no contain data on the baby names
girls_data.drop(girls_data.tail(7).index,inplace=True)

#Check if the data has been cleaned
girls_data.tail(10)
```

```python
# Remove the last 8 columns as these refer to 1996-1999 
# We will be analysing 2000-2020
girls_data.drop(girls_data.columns[43:51], axis=1, inplace=True)

# check if columns have been dropped
girls_data.head(5)
```

```python
# Take a look at the top of the data frame to assess how to clean
girls_data.head(6)
```

```python
# Need to remove the top 4 rows as they need reformatting to include informative column headers
girls_data.drop(girls_data.head(4).index,inplace=True)

# Check if the data has been cleaned
girls_data.head(10)
```

```python
# Rename columns appropriately to make up for the formatting of the original Excel spreadsheet
girls_data.rename(columns = {"Contents": "Name"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 1": "2020 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 2": "2020 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 3": "2019 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 4": "2019 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 5": "2018 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 6": "2018 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 7": "2017 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 8": "2017 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 9": "2016 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 10": "2016 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 11": "2015 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 12": "2015 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 13": "2014 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 14": "2014 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 15": "2013 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 16": "2013 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 17": "2012 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 18": "2012 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 19": "2011 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 20": "2011 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 21": "2010 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 22": "2010 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 23": "2009 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 24": "2009 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 25": "2008 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 26": "2008 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 27": "2007 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 28": "2007 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 29": "2006 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 30": "2006 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 31": "2005 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 32": "2005 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 33": "2004 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 34": "2004 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 35": "2003 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 36": "2003 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 37": "2002 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 38": "2002 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 39": "2001 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 40": "2001 Count"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 41": "2000 Rank"}, inplace = True)
girls_data.rename(columns = {"Unnamed: 42": "2000 Count"}, inplace = True)

# call data to check if the columns were renamed
girls_data.head()
```

Successfully changed the column headings

## Look at data types and change if necessary

```python
girls_data.dtypes
```

All the dtypes are objects. However, we need numeric values in order to analyse/plot graphs. Therefore, we need to change the data type to float.

```python
# This is an attempt to change 2020 Rank column to numeric (float) values so that we can analyse the data 
```

```python
girls_data['2020 Rank'] = pd.to_numeric(girls_data['2020 Rank'],errors = 'coerce')
girls_data['2020 Rank'].dtypes
```

This is a success, so apply to all rank/count columns:

```python
for i in range(1,42):
    girls_data.iloc[:, i] =pd.to_numeric(girls_data.iloc[:, i],errors = 'coerce')
    
# check if the data type changed
girls_data.dtypes
```

```python
# Check/view the data
girls_data
```

## Missing Data

Let's look at the missing data using a heatmap

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(girls_data.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("missing_data_girls.png", dpi=100)
```

There are no names are missing, but there are several missing/null data though within the ranks/counts.

```python
# This shows the % of missing data from each column
girls_data.isnull().mean().round(4) * 100 
```

```python
#Lets remove some of the data
girls_data2 =girls_data.dropna(thresh = 11) #11 = 1 + 2 * 5, where : 1(name) + 2(rank, count)* 5 (years)
girls_data2
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(girls_data2.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("missing_data_girls2.png", dpi=100)
```

```python
# This shows the % of missing data from each column
girls_data2.isnull().mean().round(4) * 100
```

The number of null data decreased.

```python
# Check to show the data in new df to see what it looks like
girls_data2
```

## Create a dataframe to hold all the RANK values for girls names in 2000 to 2020

Note : We need columns with index 0(Name), 1 (2020 Rank), 3(2019 Rank), 5,7,...41(2000 Rank) 

For that, we created a list with the desired indexes using range, list & insert:

```python
selected_rank_cols =list(range(1,42,2))
selected_rank_cols
```

```python
selected_rank_cols.insert(0,0)
selected_rank_cols
```

```python
rank_girls_data = girls_data2.iloc[:, selected_rank_cols]
rank_girls_data
```

We now need to replace the NaN values - ideally to the max value for each year (i.e. column) + 1. e.g. if the max rank in 2020 was 2345, all the names that didn't appear in that year, should have the rank 2346.


```python
# This shows the max value for each year
max_rank_year = rank_girls_data.max(axis=0).tolist()
max_rank_year
```

This removes the max name value:

```python
max_rank_year.pop(0)
```

```python
max_rank_year
```

```python
max_rank_year_plus1 = [x+1 for x in max_rank_year]
max_rank_year_plus1
```

```python
max(max_rank_year_plus1)
```

```python
rank_girls_data = rank_girls_data.fillna(5877)
rank_girls_data
```

```python
rank_girls_data.to_csv('../data/girls_rank_clean.csv')

girls_rank_clean = pd.read_csv(r'../data/girls_rank_clean.csv',index_col=[0])

girls_rank_clean
```

## Create dataframe for COUNT

We need columns with index 0(Name), 2 (2020 Count), 4(2019 Count), 6,8,...42(2000 Count). 

For that, we create a list with the desired indexes:

```python
selected_count_cols =list(range(0,43,2))
selected_count_cols
```

```python
count_girls_data = girls_data2.iloc[:, selected_count_cols] #This selects ALL (:) rows and the given columns
count_girls_data
```

## Replace NaN with zero for missing count values


```python
count_girls_data = count_girls_data.fillna(0)
count_girls_data
```

This heatmap shows there is NO missing data now.

```python
plt.figure(figsize=(10,6))
sns.heatmap(count_girls_data.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("missing_data_girls2.png", dpi=100)
```

```python
count_girls_data.to_csv('..data/girls_count_clean.csv')

girls_count_clean = pd.read_csv(r'..data/girls_count_clean.csv',index_col=[0])

girls_count_clean
```
