# Obtaining Key Stats from Baby Name Data

## Read Clean Girl CSV

```python
import pandas as pd
girls_rank_df = pd.read_csv('../data/girls_rank_clean.csv', index_col=[0])
girls_rank_df
```

## Read Clean Boy CSV

```python
boys_rank_df = pd.read_csv('../data/boys_rank_clean.csv', index_col=[0])
boys_rank_df
```

## Transpose and Clean Dataframes

```python
girls_rank_df_t = girls_rank_df.transpose()
girls_rank_df_t.head()
```

The dataframe has been transposed (rows swapped with columns and vice versa) so that we can access data on baby names via a column heading, associated with that baby name. 

```python
#Cleaning the dataframe
girls_rank_df_t.columns = girls_rank_df_t.iloc[0]
girls_rank_df_t.drop(girls_rank_df_t.head(1).index,inplace=True)
girls_rank_df_t
```

Repeat the same for boys data

```python
boys_rank_df_t = boys_rank_df.transpose()
boys_rank_df_t.columns = boys_rank_df_t.iloc[0]
boys_rank_df_t.drop(boys_rank_df_t.head(1).index,inplace=True)
boys_rank_df_t
```

## 2009 Girls Stats Function


This function should be used for girl names where the film was released in 2009 to pull statistics that will help us to answer our key questions.

```python
def girl_2009_stats(name):

        rank2008 = girls_rank_df_t[name][12]
        rank2009 = girls_rank_df_t[name][11]
        rank2010 = girls_rank_df_t[name][10]
        print(name, "STATS:\n")
        print("2008 Rank:", rank2008)
        print("2009 Rank:", rank2009)
        print("2010 Rank:", rank2010)
        
        
        pc_change_2008_to_2009 = (rank2009 - rank2008)*100/rank2008
        print("\n2008 to 2009 percentage change:", np.around(-pc_change_2008_to_2009, decimals=2), "%")
        
        pc_change_2009_to_2010 = (rank2010 - rank2009)*100/rank2009
        print("2009 to 2010 percentage change:", np.around(-pc_change_2009_to_2010, decimals=2), "%")
        
        mean_3yrs_before = np.around(np.mean(girls_rank_df_t[name][12:15]), decimals=2)
        print("\nMean rank 3 years before release:", mean_3yrs_before)
        mean_3yrs_after = np.around(np.mean(girls_rank_df_t[name][8:11]), decimals=2)
        print("Mean rank 3 years after release:", mean_3yrs_after, "\n\n")

```

```python
girl_2009_stats('TIANA') #display stats for the name Tiana from Princess and the Frog
girl_2009_stats('CORALINE') 
```
