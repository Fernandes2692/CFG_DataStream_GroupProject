# Considering statistical tests

Since we are interested in the impact of an event (movie release) on a piece of data we will need to consider statistical tests that will pair the data points and evaluate if the difference is significant. 

Consider the paired t-test, we do not make any assuptions on the distributions of the orginal data i.e. rank of baby names before and after film release but paired t-test assumes normal distributions of differences i.e. change in rank so we now evaluate the distribution of differences to see if this is a fair assumption.


# Calculating difference between ranks

We decide to use the difference in 'Mean rank 3 years before release' and 'Mean rank 3 years after release' for our tests since baby naming after film release isn't always immediate so we can get a better understanding of it's impact across the three years.

```python
Difference_Before_After_Hero = hero_data['Mean rank 3 years before release'] - hero_data['Mean rank 3 years after release']
Difference_Before_After_Hero
```

```python
Difference_Before_After_Hero.hist()
```

```python
Difference_Before_After_Villain = villain_data['Mean rank 3 years before release'] - villain_data['Mean rank 3 years after release']
Difference_Before_After_Villain
```

```python
Difference_Before_After_Villain.hist()
```

```python
Difference_Before_After_All = all_data['Mean rank 3 years before release'] - all_data['Mean rank 3 years after release']
Difference_Before_After_All
```

```python
Difference_Before_After_All.hist()
```

# Differences in rank are not normally distributed for our data so the paired t-test is not an appropriate statistical test for us to use.

# So we proceed to use the Wilcoxon signed rank test. 

Our reasons for selecting this test:
* robust against outliers & heavy tail distributions
* does not require a special distribution of the data

We are testing our data based on the following null (H0) and alternative (H1) hypothesese. 
Put more simply, for a two tail test we have:

* H0: no change
* H1: a change 

and for a one tail test we have:

* H0: no change
* H1: an increase

or 

* H0: no change
* H1: a decrease

depending on which tail of the distribution we are interested in.


We are testing all of our data at the 5% significance level meaning that if our test gives a p value < 0.05 we will reject our H0 in favour of H1.

Note that the baby name here will depend on which dataset we are using for example for the heroes data we will be referring to effect that the hero names had on rank of baby names, the same for the villians and then we will also look at both combined.

# We apply the Wilcoxon test to the Hero Data

Note because we are looking at the difference, where this is the mean rank for 3 years before the film subtract the mean for 3 years after the film then this difference, being positive corresponds to the baby name increasing in popularity as it is decreasing in rank after the film. 

# We import the test function

```python
from scipy.stats import wilcoxon
```

## We conduct a two tail test on the hero data to test if the difference (in rank) is different to 0 i.e. the baby name changes in popularity after the film is released with a hero with said name

We formulate this as the following null and alternative hypotheses: 
* H0: That the rank of a baby name remains unchanged before and after a film released with a hero with this name.
* H1: That the rank of a baby name changes from before to after a film is released with a hero with this name.

```python
d1 = Difference_Before_After_Hero
w, p = wilcoxon(d1)
w, p
```

## We conduct a one tail test on the hero data to test if the difference (in rank) is greater than 0 i.e. the baby name becomes more popular after the film is released with a hero with this name

We formulate this as the following null and alternative hypotheses: 
* H0: That the rank of a baby name remains unchanged before and after a film released with a hero with this name.
* H1: That the rank of a baby name decreases (i.e. becomes more popular) from before to after a film is released with a hero with this name.

```python
d1a = Difference_Before_After_Hero
w, p = wilcoxon(d1a, alternative = 'greater')
w, p
```

# Conclusion for the hero data

In our first Wilcoxon test (two-tailed test), we see that p=0.0054931640625<0.05 and so we reject H0 in favour of H1. So, there is enough evidence to say that the rank of a baby name changes before and after a film is released with a hero with said name.

In our second Wilcoxon test (one-tailed test), p=0.0054931640625<0.01 so we have evidence to say that even at the 1% significance level that a baby name increases in popularity after a film is released with a hero of said name.

# Now we test the villain data

## We conduct a two tail test on the villain data to test if the difference (in rank)  is different to 0 i.e. the baby name changes in popularity after the film is released with a villain with this name

We formulate this as the following null and alternative hypotheses: 
* H0: That the rank of a baby name remains unchanged before and after a film released with a villain with this name.
* H1: That the rank of a baby name changes from before to after a film is released with a villain with this name.

```python
d2 = Difference_Before_After_Villain
w, p = wilcoxon(d2)
w, p
```

## We conduct a one tail test on the villain data to test if the difference (in rank) is less than 0 i.e. the baby name decreases in popularity after the film is released with a villain with this name

* H0: That the rank of a baby name remains unchanged before and after a film released with a villain with this name.
* H1: That the rank of a baby name increases (i.e. becomes less popular) from before to after a film is released with a villain with this name.

```python
d2a = Difference_Before_After_Villain
w, p = wilcoxon(d2a, alternative = 'less')
w, p
```

# Conclusion for the villain data 

In our first Wilcoxon test (two-tailed test), we see that p=0.125>0.05 and so we do not have sufficent evidence to reject H0. So, there is no evidence to say that the rank of a baby name changes before and after a film is released with a villain with said name. 

In our second Wilcoxon test (one-tailed test), p=0.0625>0.1 so we have (weak) evidence, as its only at the 10% significance level, to show a the baby name decreases in popularity after a film is released with a villain of said name. However, working at the 5% level we would still reject this and accept H0 saying that we saw no significant change in rank of baby names when a film was released with a villain of said name.

To improve this test we would increase our sample size, but due to the nature of the films from the family genre not all of the top 3 films from each year had villains.
