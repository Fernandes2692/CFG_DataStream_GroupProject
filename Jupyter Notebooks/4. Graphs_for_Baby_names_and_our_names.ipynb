# Read the cleaned data files

```python
import pandas as pd
girls_rank_df = pd.read_csv('C:\\Users\sukic\Documents\CFG_Nanodegree\Data_Stream\Project//girls_rank_clean.csv', index_col=[0])
girls_rank_df
#save girls data to df
```

```python
boys_rank_df = pd.read_csv('C:\\Users\sukic\Documents\CFG_Nanodegree\Data_Stream\Project//boys_rank_clean.csv', index_col=[0])
boys_rank_df
#save boys data to df
```

## Transpose the rank dataframes

```python
girls_rank_df_t = girls_rank_df.transpose()
girls_rank_df_t.head()
#transpose the data (swap rows & columns) so that you can access names through column headings
```

Below we have set the column headings to the girls' names and dropped the first row containing the baby names

```python
girls_rank_df_t.columns = girls_rank_df_t.iloc[0]
girls_rank_df_t.drop(girls_rank_df_t.head(1).index,inplace=True)
girls_rank_df_t
#quick clean up of data
```

Do the same for the boys rank dataframe:

```python
boys_rank_df_t = boys_rank_df.transpose()
boys_rank_df_t.columns = boys_rank_df_t.iloc[0]
boys_rank_df_t.drop(boys_rank_df_t.head(1).index,inplace=True)
boys_rank_df_t
```

## Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

To plot graphs, a list including the years 2020 down to 2000 would be useful. Here is a for loop to create such list:

```python
year_list_1 = []
for i in range(21):
    year = 2020 - i
    year_list_1.append(str(year))
print(year_list_1) 
```

To explore the data we have, let's look at the top 5 girls names from 2020 and plot their rank for the past 10 years

```python
rank2020_girls_df = girls_rank_df.sort_values(by = ['2020 Rank']).head(5)
rank2020_girls_df
#check that we have accessed the top five baby girls names from 2020
```

```python
top5_2020_girls_name = rank2020_girls_df['Name']
top5_girls_2020_list = top5_2020_girls_name.tolist()
top5_girls_2020_list
#assign these girls names to a list for future plotting
```

```python
top5_girls_2020_list[0]
#check that Olivia appears as the top ranking name
```

```python
rank_1_girl =top5_girls_2020_list[0]
rank_2_girl =top5_girls_2020_list[1]
rank_3_girl =top5_girls_2020_list[2]
rank_4_girl =top5_girls_2020_list[3]
rank_5_girl =top5_girls_2020_list[4]

fig, ax = plt.subplots(figsize=(25,18))    
plt.plot(year_list_1, girls_rank_df_t[rank_1_girl], label = rank_1_girl, linewidth = 19, color="pink")
plt.plot(year_list_1, girls_rank_df_t[rank_2_girl], label = rank_2_girl, linewidth = 15)
plt.plot(year_list_1, girls_rank_df_t[rank_3_girl], label = rank_3_girl, linewidth = 11)
plt.plot(year_list_1, girls_rank_df_t[rank_4_girl], label = rank_4_girl, linewidth = 9)
plt.plot(year_list_1, girls_rank_df_t[rank_5_girl], label = rank_5_girl, linewidth = 5)


cfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Helvetica'}
plt.legend(loc="lower left")
plt.title("\nTop 5 Ranking Girls' Names in 2020,\n Over the Past 10 Years\n", fontsize = 50, **hfont)
plt.xlabel('\nYear', fontsize=40, **hfont)
plt.ylabel('Rank\n', fontsize=40, **hfont)
plt.xticks(fontsize=30, rotation = 70, **cfont)
plt.yticks(fontsize=30, **cfont)
ax.legend(fontsize=30)
plt.gca().invert_yaxis()
plt.xlim([10, 0])
plt.ylim([20, 0])

plt.savefig('C:\\Users\sukic\Documents\CFG_Nanodegree\Data_Stream\Project//Top5RankingGirlsNames.png',dpi=300)
plt.show() 
```
